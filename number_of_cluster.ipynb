```python
import pandas as pd
import matplotlib.pyplot as plt
import csv
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans

#loading dataset
import pandas as pd
data = pd.read_csv('/content/drive/MyDrive/cluster_data.csv')
plt.show()

#drop ID column we dont't use this column
df=data.drop(['LCLid'],axis=1)

#scaling the dataset
mms = MinMaxScaler()
mms.fit(df)
data_transformed = mms.transform(df)

#convert to dataframe
data_transformed = pd.DataFrame(data_transformed)


#elbow method to minimise WSS(within cluster sum of sqr)
Sum_of_squared_distances = []
K = range(1,15)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(data_transformed)
    Sum_of_squared_distances.append(km.inertia_)

#plotting the elbow curve by k
plt.plot(K,Sum_of_squared_distances,'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.axis([0,6,0,35000])
plt.title('elbow method for optional k')
plt.show()

#
```

```python
from yellowbrick.cluster import KElbowVisualizer
model = KMeans()
# k is range of number of clusters.
visualizer = KElbowVisualizer(model, k=(2,10),metric='silhouette', timings= False)
visualizer.fit(data_transformed)        # Fit the data to the visualizer
```
